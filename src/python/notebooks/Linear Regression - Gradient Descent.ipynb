### Import

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

### Parameters

```python
nPoints = 100
nDim = 2
learningRate = 0.05
noiselevel = 0.0025
nEpochs = 500
basecoef = np.random.random((nDim))
```

### Generate Data

```python
X = np.random.random((nPoints,nDim))
X[:,-1] = 1
y = np.dot(X,basecoef)
n = np.random.normal(0,1,nPoints) * noiselevel
y = y + n
```

### Plot Data

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:,0]+X[:,1],y);
```

### Gradient Descent

```python
cost = []
coef = np.random.random((nDim))
for i in range(nEpochs):
    h = np.dot(X,coef)
    error = h - y
    cost.append(np.sum(error**2))
    gradient = np.dot(X.T,error) / y.shape[0]
    coef = coef - (gradient*learningRate)
```

### Plot Cost

```python
plt.figure(figsize=(8,8))
plt.plot(cost)
```

### Calculate Rsquare

```python
ypred = np.dot(X,coef)
SSE = np.sum((y - ypred)**2)
SST = np.sum((y - np.mean(y))**2)
rsquare = 1 - (SSE / SST)
print rsquare
```
