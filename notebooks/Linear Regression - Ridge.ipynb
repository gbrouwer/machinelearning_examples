### Import

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv
```

### Parameters

```python
nPoints = 100
nDim = 10
learningRate = 0.05
noiselevel = 0.25
nEpochs = 500
basecoef = np.random.random((nDim))
```

### Generate Data

```python
X = np.random.random((nPoints,nDim))
X[:,-1] = 1
y = np.dot(X,basecoef)
n = np.random.normal(0,1,nPoints) * noiselevel
y = y + n
```

### Plot Data

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:,0]+X[:,1],y);
```

### Ridge Regression

```python
k = 0.01
I = np.eye(nDim)
B = np.dot(np.dot(inv(np.dot(X.T,X) + k*I),X.T),y)
```

### Choose K

```python
kvals = 10**np.linspace(-12,12,23)
betas = []
for k in kvals:
    B = np.dot(np.dot(inv(np.dot(X.T,X) + k*I),X.T),y)
    betas.append(B)
betas = np.array(betas)
```

### Plot

```python
plt.plot(betas);
```
